```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd

logdir = "/home/jovyan/logs"
run = "PPO-Var0-(count-time)-1_1"  # or whichever run

ea = EventAccumulator(f"{logdir}/{run}")
ea.Reload()

# Get ep_rew_mean
events = ea.Scalars("rollout/ep_rew_mean")
df = pd.DataFrame([(e.step, e.value) for e in events], columns=["timestep", "ep_rew_mean"])

# Mean reward over last 10% of training
last_10pct = df[df["timestep"] > df["timestep"].max() * 0.9]
print(last_10pct["ep_rew_mean"].mean(), last_10pct["ep_rew_mean"].std())
```
