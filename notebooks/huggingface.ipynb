### Download Models

```python
from huggingface_hub import snapshot_download

repo_id = "boettiger-lab/rl4eco"
local_dir = "./models"

files_to_download = [
    "rl4greencrab/model/twoActEnv/TQC_Var_2obs.zip",
    "rl4greencrab/model/twoActEnv/TQC_Var_Nobs.zip",
    "rl4greencrab/model/twoActEnv/TQC_noVar_2obs.zip",
    "rl4greencrab/model/twoActEnv/TQC_noVar_Nobs.zip",
]

downloaded_path = snapshot_download(
    repo_id=repo_id,
    repo_type="model",
    allow_patterns=files_to_download,
    local_dir=local_dir,
)

print("Downloaded into:", downloaded_path)
```
